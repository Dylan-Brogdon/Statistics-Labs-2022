```python
import matplotlib.pyplot as plt
```

```python
labels = ['Frogs', 'Dogs', 'Cats', 'Birds']

sizes = [15,30,45,10]

explode = (0,0.1,0,0)

fig1,ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal');
```

```python
#We created a pie chart using multiple pieces of code
#The Labels are labels
#Sizes are the percentages
#Explode extrudes a peice of the chart
#Shadow is true and the start angle is 90
```

```python
labels = ['Burgers', 'Fries', 'Soft-Drinks', 'Shakes']

sizes = [60,25,10,5]

explode = (0.1,0,0,0)

fig1,ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=10)
ax1.axis('equal');
```

```python
heads = 'https://random-ize.com/coin-flip/us-quarter/us-quarter-front.jpg'
tails = 'https://random-ize.com/coin-flip/us-quarter/us-quarter-back.jpg'
```

```python
def show_img(img):
  display(HTML('<img src=' + img + '>'))
```

```python
show_img(heads)
```

```python
show_img(tails)
```

```python
def flip_coin():
  if np.random.random() > 0.5:
    show_img(heads)
  else:
    show_img(tails)
```
